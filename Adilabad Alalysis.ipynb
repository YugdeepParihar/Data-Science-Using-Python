```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/28_Adilabad census.csv')
```

```python

df.plot(kind='scatter', x='TRU', y='TOT_P')

plt.title('total population on each level')

plt.ylabel('Total population')
```

```python
sns.histplot(data=df,x="CD Block")
```

```python
non_work_man = (df['NON_WORK_M'].sum()/df['NON_WORK_P'].sum()).tolist()
non_work_woman = (df['NON_WORK_F'].sum()/df['NON_WORK_P'].sum()).tolist()

lst = [non_work_man , non_work_woman]
gender = ['non_work_man' , 'non_work_woman']
plt.pie(lst , labels = gender , autopct="%.1f%%",shadow=True,textprops={"color":"b"})
plt.title("Non Working population",color = "white")
plt.show()
```
